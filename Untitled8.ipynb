```python
def verificar_acesso(cargo, dia):
    dias_uteis = ["segunda-feira", "terça-feira", "quarta-feira", "quinta-feira", "sexta-feira"]
    cargo = cargo.lower()
    dia = dia.lower()

    if cargo == "gerente":
        return "Acesso permitido"
    elif cargo in ["analista", "estagiário"]:
        if dia in dias_uteis:
            return "Acesso permitido"
        else:
            return "Acesso negado"
    else:
        return "Cargo inválido"

# Exemplo de uso:
if __name__ == "__main__":
    cargo_input = input("Digite o cargo do funcionário: ")
    dia_input = input("Digite o dia da semana: ")

    resultado = verificar_acesso(cargo_input, dia_input)
    print(resultado)
```

```python
def pode_acessar(cargo, dia_semana):
    cargo = cargo.strip().lower()
    dia_semana = dia_semana.strip().lower()

    dias_uteis = {"segunda-feira", "terça-feira", "quarta-feira", "quinta-feira", "sexta-feira"}

    if cargo == "gerente":
        return True
    elif cargo in {"analista", "estagiário"} and dia_semana in dias_uteis:
        return True
    else:
        return False


def entrada_usuario():
    cargo = input("Digite o cargo do funcionário: ")
    dia_semana = input("Digite o dia da semana: ")

    if pode_acessar(cargo, dia_semana):
        print("Acesso Permitido.")
    else:
        print("Acesso Negado.")


# Executar o programa
if __name__ == "__main__":
    entrada_usuario()
```

```python
import random

# Tamanho do tabuleiro
TAMANHO = 5

# Criar o tabuleiro vazio
def criar_tabuleiro():
    return [["~" for _ in range(TAMANHO)] for _ in range(TAMANHO)]

# Mostrar o Tabuleiro
def mostrar_tabuleiro(tabuleiro, ocultar=False):
    print("  " + " ".join(str(i) for i in range(TAMANHO)))
    for i, linha in enumerate(tabuleiro):
        print(str(i) + " " + " ".join("~" if ocultar and cel == "N" else cel for cel in linha))

# Posicionar o navio
def posicionar_navio(tabuleiro, tamanho_navio=3):
    orientacao = random.choice(["H", "V"])
    colocado = False

    while not colocado:
        if orientacao == "H":
            linha = random.randint(0, TAMANHO - 1)
            coluna = random.randint(0, TAMANHO - tamanho_navio)
            # Verifica se espaço está livre
            if all(tabuleiro[linha][coluna + i] == "~" for i in range(tamanho_navio)):
                for i in range(tamanho_navio):
                    tabuleiro[linha][coluna + i] = "N"
                colocado = True
        else:  # Vertical
            linha = random.randint(0, TAMANHO - tamanho_navio)
            coluna = random.randint(0, TAMANHO - 1)
            if all(tabuleiro[linha + i][coluna] == "~" for i in range(tamanho_navio)):
                for i in range(tamanho_navio):
                    tabuleiro[linha + i][coluna] = "N"
                colocado = True

# Jogada do jogador
def jogada(tabuleiro, linha, coluna):
    if tabuleiro[linha][coluna] == "N":
        tabuleiro[linha][coluna] = "X"
        print("💥 Acertou um navio!")
        return True
    elif tabuleiro[linha][coluna] == "~":
        tabuleiro[linha][coluna] = "O"
        print("🌊 Água!")
    else:
        print("⚠️ Já tentou essa posição.")
    return False

# Verifica se todos os navios foram afundados
def verificar_fim(tabuleiro):
    for linha in tabuleiro:
        if "N" in linha:
            return False
    return True

# Função principal
def jogar():
    print("=== 🛳️ BATALHA NAVAL ===")
    tabuleiro = criar_tabuleiro()
    posicionar_navio(tabuleiro)

    tentativas = 0
    while not verificar_fim(tabuleiro):
        mostrar_tabuleiro(tabuleiro, ocultar=True)
        try:
            linha = int(input("Escolha a linha (0-4): "))
            coluna = int(input("Escolha a coluna (0-4): "))
            if 0 <= linha < TAMANHO and 0 <= coluna < TAMANHO:
                jogada(tabuleiro, linha, coluna)
                tentativas += 1
            else:
                print("❌ Posição inválida. Tente novamente.")
        except ValueError:
            print("⚠️ Digite números válidos.")

    print("\n🎯 Parabéns! Você afundou todos os navios em", tentativas, "tentativas!")
    mostrar_tabuleiro(tabuleiro)

# Executar o jogo
if __name__ == "__main__":
    jogar()
```

```python
def exibir_tabuleiro(tabuleiro):
    print("\n")
    print(f" {tabuleiro[0]} | {tabuleiro[1]} | {tabuleiro[2]}")
    print("---+---+---")
    print(f" {tabuleiro[3]} | {tabuleiro[4]} | {tabuleiro[5]}")
    print("---+---+---")
    print(f" {tabuleiro[6]} | {tabuleiro[7]} | {tabuleiro[8]}")
    print("\n")


def verificar_vitoria(tabuleiro, jogador):
    combinacoes_vitoria = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # linhas
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # colunas
        [0, 4, 8], [2, 4, 6]              # diagonais
    ]
    for combinacao in combinacoes_vitoria:
        if all(tabuleiro[i] == jogador for i in combinacao):
            return True
    return False


def jogo_da_velha():
    tabuleiro = [" "] * 9
    jogador_atual = "X"
    jogadas = 0

    print("=== JOGO DA VELHA ===")
    print("Posições do tabuleiro:")
    print(" 1 | 2 | 3 ")
    print("---+---+---")
    print(" 4 | 5 | 6 ")
    print("---+---+---")
    print(" 7 | 8 | 9 ")
    print("======================\n")

    while True:
        exibir_tabuleiro(tabuleiro)

        try:
            posicao = int(input(f"Jogador {jogador_atual}, escolha uma posição (1-9): ")) - 1

            if posicao < 0 or posicao > 8:
                print("❌ Posição inválida! Escolha entre 1 e 9.")
                continue

            if tabuleiro[posicao] != " ":
                print("⚠️ Essa posição já está ocupada!")
                continue

        except ValueError:
            print("❌ Por favor, digite um número válido.")
            continue

        tabuleiro[posicao] = jogador_atual
        jogadas += 1

        # Verifica vitória
        if verificar_vitoria(tabuleiro, jogador_atual):
            exibir_tabuleiro(tabuleiro)
            print(f"🏆 Jogador {jogador_atual} venceu!")
            break

        # Verifica empate
        if jogadas == 9:
            exibir_tabuleiro(tabuleiro)
            print("🤝 Empate!")
            break

        # Alterna jogador
        jogador_atual = "O" if jogador_atual == "X" else "X"

if __name__ == "__main__":
    jogo_da_velha()
```
